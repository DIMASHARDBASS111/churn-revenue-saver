## Очистка данных (Data Cleaning)

В этом ноутбуке выполняется очистка датасета Telcom Customer Churn:
- обработка пропусков,
- приведение типов,
- удаление проблемных строк,
- сохранение чистых данных в `processed/clean_churn.csv`.

### 1. Загрузка исходного датасета

Проверяем структуру таблицы и типы данных.

```python
import pandas as pd

df = pd.read_csv("../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
df.info()
```

### 2. Обнаружение некорректных значений TotalCharges

Поле `TotalCharges` имеет тип object, хотя должно быть числовым.
Проверяем количество строк, где поле пустое.

```python
(df['TotalCharges'] == ' ').sum()
```

### 3. Приведение TotalCharges к числовому типу

Используем `pd.to_numeric` с параметром `errors='coerce'`,
что заменяет некорректные значения на NaN.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.info()
```

```python
df['TotalCharges'].isna().sum()
```

### 4. Удаление строк с пропущенным TotalCharges

Итого 11 строк.
Для анализа churn от них можно избавиться.

```python
df = df.dropna(subset = 'TotalCharges')
df = df.reset_index(drop = True)
```

```python
df.info()
```

### 5. Сохранение очищенных данных

Сохраняем чистый датасет в `processed/clean_churn.csv`.

```python
df.to_csv("../data/processed/clean_churn.csv", index = False)
```
